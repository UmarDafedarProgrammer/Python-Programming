##### list

list is a ordered, mutable collection of reference to objects.

- Order is preserved
- Indexing starts from 0
- Every element has a position
- Does not store actual objects, instead stores references to objects

list = [ref1->10, ref2->"Python", ref3->36.89]

- Dynamic array
- Mutable
- Ordered
- Heterogeneous
- Supports slicing

```python
l = [1,2,3,4,5,6,7]
for i in range(len(l)):
    print(i,"-",l[i],"-",id(l[i]))
print(id(l))
```

```python
l1 = l.copy()
for i in range(len(l)):
    print(i,"-",l[i],"-",id(l[i]))
    print(i,"-",l1[i],"-",id(l1[i]))
print(id(l))
print(id(l1))

# new reference for list. However, the inside elements are shallow copied
```

```python
# Deep Copy
import copy

del l1
del l
l = [1,2,3,4,[5,6,7]]
for i in range(len(l)):
    print(i,"-",l[i],"-",id(l[i]))
print(id(l))

l1 = copy.deepcopy(l)
for i in range(len(l)):
    print(i,"-",l[i],"-",id(l[i]))
    print(i,"-",l1[i],"-",id(l1[i]))
print(id(l))
print(id(l1))
```
