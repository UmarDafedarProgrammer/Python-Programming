#### Floating Point Numbers
1. To represent real numbers
2. Numbers with decimal positions

```python
# Computers store floating point numbers in scientific notations
# (-1)^sign * 1.mantissa * (2)^exponent
# Sign(0-Positive, 1-Negative) | Exponent | Mantissa (Fraction)
```

##### Binary Approximation
1. Some decimal numbers can not be represented exactly in binary - so they are represented using Binary Approximation

#### Special Floating Point Numbers
1. inf -> infinity
2. -inf -> Negative Infinity
3. NaN -> Not a Number

```python
f = float() # using constructor and default value of 0.0
f
```

```python
f1 = float(8)
f1
```

##### Type of the floating point class and its object size
1. <class 'float'>
2. type(object)
3. id(object)
4. sys.getsizeof(object)

```python
print(type(f1))
```

```python
import sys
print("Size of floating point object: ",sys.getsizeof(f1)," bytes")
```

```python
print(id(f1))
```

```python
print(0.1+0.2) # Binary approximation 

# Because internally:

# 0.1 ≈ 0.10000000000000000555...
# 0.2 ≈ 0.20000000000000001110...
```

```python
# To see the real stored values in the memory
print(format(0.1,".20f"))
print(format(0.2,".20f"))
# print(type(print(0.1)))
```

```python
import math

print(float("inf"))
print(float("-inf"))
print(float("nan"))
```

```python
# validate
print(math.isnan(float("nan")))
print(math.isinf(float("inf")))
print(math.isinf(float("-inf")))
```
