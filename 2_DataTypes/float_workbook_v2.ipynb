```python
#Creating a float object

f = 10.0
f1 = float() # default constructor
f2 = float(3.222) # Parameterised constructor
f3 = float("3.14") # Type casting

# f4 = float("abc") ValueError: could not convert string to float: 'abc'

print(f, f1, f2, f3, sep="\n")
```

```python
import os
import sys

# Size includes Python object overhead. actual size is 8 bytes
print(sys.getsizeof(f)) 
print(sys.getsizeof(f1))
print(sys.getsizeof(f2))
print(sys.getsizeof(f3))
```

```python
#Checking for type

print(type(f))
```

```python
isinstance(f,float) # Returns true if the object is float
```

```python
## Rounding

# FOllows Bankers rounding - nearest even number
# When a number is exactly halfway between two values, it rounds to the nearest EVEN number. (.5)

print(round(3.5))
print(round(2.5))
print(round(2.55))
print(round(3.45))
print(round(3.4533555,3))
```

```python
# Floor and Ceil
import math
print(math.floor(3.6))
print(math.ceil(3.6))
print(math.floor(-3.6))
print(math.ceil(-3.6))
```

```python
#Truncation
print(int(3.4))
print(int(-3.4))
```

```python
#absolute value
print(abs(3.4))
print(abs(-3.4))
```

```python
# Power and Square Root
print(pow(2.0, 3))      # 8.0
print(math.sqrt(16.0))  # 4.0
```

```python
#Modulo with float
5.5%2
```

```python
# Floats are stored as binary approximations
# As float numbers can not be stored as exact binaries. so they are stored in nearest binary approximations
# That is why, comparison operator might not work for below
0.1 + 0.2 == 0.3
```

```python
print(0.1+0.2)
```

```python
import math
math.isclose(0.1+0.2,0.3)
```

```python
from decimal import Decimal

val = Decimal("0.1")+Decimal("0.2")

print(val)

val == Decimal("0.3")
```

```python
# round(value, precision)
# Defaults to rounding to the nearest integer.
# Uses banker's rounding for .5 values (round half to even).
# Specify precision to control decimal places.

round(3.456)
```

```python
# Bankers rounding
print(round(3.5))
print(round(2.5))
```

```python
print(round(3.45277,2))
print(round(3.45577,2))
```

```python
print(round(-3.45277,2))
print(round(-3.45577,2))
```

```python
print(round(3.45277,-2))
print(round(3.45577,-2))
```

```python
print(round(93.45277,-3)) # 1000 nearest  10**3
print(round(93.45577,-2)) # 100 nearest 10**2
```

```python
# Bankers rounding
round(50, -2)    # 0.0
round(150, -2)   # 200.0
```
