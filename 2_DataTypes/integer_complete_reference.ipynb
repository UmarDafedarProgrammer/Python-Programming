```python
### Arithmetic Operations
i1 = 90
i2 = 7

print(i1+i2) # addition
print(i1-i2) # subtration
print(i1*i2) # Multiplication
print(i1/i2) # Division -> Float (Quotient)
print(i1//i2) # Floor division -> Whole Number (Quotient) 
print(i1%i2) # Remainder (modulus)
print(i1**i2) ## Exponential
```

```python
### Comparison Operations
print(i1==i2) # Equal To
print(i1!=i2) # Not equal to
print(i1<i2) # Less than
print(i1>i2) # Greater than
print(i1<=i2) # Less than or equal to
print(i1>=i2) # Greater than or equal to
```

```python
### Bitwise Operations
print("90->",bin(i1)) # 90 -> 1011010
print("7->",bin(i2)) # 7 -> 0b111

print()
print(i1&i2) # Bitwise And
print(i1|i2) # Bitwise OR
print(i1^i2) # Bitwise XOR
print(~i1)   # Bitwise NOT ~i1 = -(i1+1)

print(bin(i1<<3)) # Left shift operator
print(bin(i1>>3)) # Right shift operator
```

```python
### Compound assignment Operators
i1 = 90
i3 = i1 # Assignment operator
print(i3,i1,i2)
i1 += i2 # Add and assign
print(i1,i2)
i1 -= i2 # Subtract and assign
print(i1,i2)
i1 *= i2 # Multiply and assign
print(i1,i2)
i1 /= i2 # Quotient and Assign
print(i1,i2)
i1 //= i2 # Floor division and Assign
print(i1,i2)
i1 %= i2 # Modulus(Remainder) and Assing
print(i1,i2)
i1 **= i2 # Exponential and Assign
print(i1,i2)

i1 = round(i1)
print(i1)
```

```python
i1 &= i2 # i1 & i2 and assign to i1
print(i1, i2)
i1 |= i2 # i1|i2 and assign to i1
print(i1,i2)
i1 ^= i2 # i1^i2 and assing to i1
print(i1,i2)
i1 <<= 2 # i1<<2 and assing to i1
print(i1,i2)
i1 >>= 1 # i1>>1 and assign to i1
print(i1,i2)
```

```python
### Built int Functions
i1 = -98
i2 = 5

print(abs(i1)) #absolute value of i -> Positive integer
print(round(98.9898)) # Rounds to integer value
print(pow(2,9)) # 2 to the power 9
print(divmod(12,9)) # Returns (a//b(quotient), a%b(remainder))

print(bin(19)) # binary representation (0b)
print(hex(19)) # hexadecimal equivalent (0x)
print(hex(15))
print(oct(19)) # Octal equivalent (0o)
print(oct(15))
```

```python
### Internal operator methods (dunder methods)
# Dunder methods -> Double + Underscore methods
i1 = 90
print(i1.__add__(10)) # +
print((5).__add__(10)) 

print((15).__sub__(10)) # -
print(i1.__sub__(10))
print(i1.__mul__(5)) # *
print(i1.__truediv__(4)) # /
print(i1.__floordiv__(4)) # //
print(i1.__mod__(4)) # %
print(i1.__pow__(4)) # **
print(i1.__lshift__(2)) # <<
print(i1.__rshift__(2)) # >>
print(i1.__float__()) # float()
print((-91).__abs__()) # abs()
print((-92.99).__int__()) # int()
print(i1.__and__(i2)) # &
print(i1.__or__(i2)) # |
print(i1.__xor__(i2)) # ^
print(i1.__neg__()) # ~
```

```python
# Dunder methods for comparison operators
print(i1.__eq__(i2)) # ==
print(i1.__lt__(i2)) # <
print(i1.__gt__(i2)) # >
print(i1.__le__(i2)) # <=
print(i1.__ge__(i2)) # >=
print(i1.__ne__(i2)) # !=
```

```python
# Method	Operator	Behavior
# __add__	+	Creates a new object
# __iadd__	+=	Modifies existing object if possible
```

```python
a = 5
a+=5
# a = a.__iadd__(3) For immutable types like int, Python does not define __iadd__
# Python internally does: for a += 5
# 1️⃣ Try a.__iadd__(3) (works only for mutable objects and for immutable objects this function is not defined)
# 2️⃣ If not available → use a.__add__(3)
# 3️⃣ Assign result back to a
print(a)
```

```python
a = [1,2]

print(a.__iadd__([3]))
# Quick Comparison
# Type	__iadd__ exists?	Behavior
# int	❌ No	Falls back to __add__
# float	❌ No	Falls back to __add__
# list	✅ Yes	Modifies same object
# set	✅ Yes	Updates in place
# dict	❌ No (+= not supported)	
```
