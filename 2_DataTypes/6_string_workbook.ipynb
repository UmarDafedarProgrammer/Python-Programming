##### String interning

```python
a = "hello"
b = "hello"

print(id("hello"), id(a), id(b))
```

```python
a == b
```

```python
a is b
```

```python
sys.intern("long_string")
a1 = "This is just one of the case to validate if the interning works for long strings"
a2 = "This is just one of the case to validate if the interning works for long strings"

a1 is a2
```

```python
a1 = "This is just one of the case to validate if the interning works for long strings"
a2 = a1

a1 is a2
```

```python
a1 == a2
```

```python
s = "This is just one of the case to validate if the interning works for long strings"

a1 = sys.intern(s)
a2 = sys.intern(s)

a1 is a2
```

#####
Python does automatically apply interning for the small strings for the optimization as the strings are immutable. However, for the long strings, it is not applied automatically.

We will have to use var = sys.intern("string") for forcing Python to store one canonical copy

and var1 = var2 (applying the reference to var2)

#### Encoding and Decoding. Bytes

```python
S = "H" # Considered as Unicode character
print(S)
print(type(S))
print(S[0])
print(len(S))

B = b"H" # it is considered as Byte
print(B)
print(type(B))
print(B[0]) 
print(len(B))
```
