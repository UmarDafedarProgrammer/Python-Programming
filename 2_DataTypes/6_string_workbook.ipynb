1. Immutable sequence of Unicode characters

```python
import os
import sys

s = "python"

print(s)
print(sys.getsizeof(s))
print(type(s))
print(id(s))
```

```python
for i in range(len(s)):
    print(i)

print()

for i in range(len(s)):
    print(s[i])

print()

for i in s:
    print(i)
```

##### 5. Indexing and Slicing

```python
print(s)
print(s[0])
print(s[:])
print(s[:4])
print(s[::2])
print(s[::1]) 
print(s[::-1]) # Reverse Step 1
print(s[::-2]) # Reverse skip step 2
print(s[1:5:2])
```

##### Split Join Partition

```python
s1 = "Helllo World. I am Working On Python"
```

```python
s_list = s1.split(" ") # Split is generating the list
s_list
```

```python
s2 = "-".join(s_list) # Takes the input as list and generates the string with separator between each element
```

```python
type(s2.partition("-")) # Returns tuple
s2.partition("-")
```
