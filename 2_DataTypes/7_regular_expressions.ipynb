##### Regular Expression

```python
import re
```

```python
S = "Sam2 4ple 7 8 8 9"
```

```python
l1 = re.findall(r'\d',S)
print(l1)
```

```python
l2 = re.findall(r'\D',S)
print(l2)
```

```python
l3 = re.findall(r'\D+',S)
print(l3)
```

```python
re.findall(r"^\d+", "123abc")   # ['123']
```

```python
re.findall(r"\d+$", "abc123")   # ['123']
```

```python
re.findall(r"\d", "abc123def456") 
```

```python
re.findall(r"\d+$", "abc123def456") 
```

```python
re.findall(r"\d+", "abc123def456")  
```

```python
re.findall(r"^\d+", "123abc123def456")   
```

```python
re.findall(r"\d+$", "123abc123def456")   
```

```python
re.findall(r"^\d", "123abc123def456")
```

```python
re.findall(r'\d',"123abc345cdef")
```

```python
re.findall(r'(ab)',"123abcabab345cdef")
```

```python
re.findall(r"(\d+)-(\d+)","123abcabab345-67cdef")
```

```python
re.findall(r"(?:ab)+", "123abcabab345cdef")
```

```python
re.findall(r"a|b", "123abcabab345cdef")
```

```python
re.findall(r"\d(?=px)", "10px 20px")
```

```python
re.findall(r"(?<=₹)\d+", "₹100 ₹200")
```

```python
re.findall(r"\.", "a.b.c")  # ['.', '.']
```

```python
re.sub(r"\d", "#", "a1b2")  # 'a#b#'
```

```python
re.split(r"[,\s]+", "a, b  c")
```
