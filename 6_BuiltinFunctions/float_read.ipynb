# 🧠 Floating Point Numbers (float) — Reference Notes

## 🔹 What is a Float?
A floating point number (`float`) represents **real numbers (numbers with decimals)** in Python.

**Examples:**
- 3.14  
- -0.001  
- 2.0  
- 1e3 (scientific notation for 1000.0)

---

## 🔹 Why "Floating" Point?
The decimal point can **move (float)** depending on representation.

**Example:**
- 123.45  
- 1.2345 × 10²  
- 0.012345 × 10⁴  

All represent the same value.

---

## 🔹 Internal Representation
- Stored using the **IEEE 754 standard**
- Represented in **binary (base 2)**
- Leads to **approximation (not exact values)**

---

## 🔹 Precision Limitation ⚠️
Not all decimal numbers can be represented exactly in binary.

**Example:**
0.1 + 0.2 = 0.30000000000000004


👉 Due to **binary rounding error**

---

## 🔹 Key Properties
- Floats are **approximate values**
- Precision is about **15–17 decimal digits**
- Arithmetic may introduce **small errors**
- Suitable for **scientific calculations**

---

## 🔹 Comparing Floats ⚠️

❌ Avoid:
0.1 + 0.2 == 0.3

✅ Use:
mport math
math.isclose(0.1 + 0.2, 0.3)


---

## 🔹 Type Conversion
- `float(5)` → 5.0  
- `int(3.9)` → 3 (**truncates, not rounds**)

---

## 🔹 Common Functions
- `round(x, n)` → rounds to n decimals  
- `abs(x)` → absolute value  
- `pow(x, y)` → exponentiation  
- `divmod(a, b)` → quotient & remainder  

---

## 🔹 Special Float Values
- `float("inf")` → Infinity  
- `float("-inf")` → Negative Infinity  
- `float("nan")` → Not a Number  

---

## 🔹 Scientific Notation
- `1e3` → 1000.0  
- `2.5e-3` → 0.0025  

---

## 🔹 Common Mistakes ❌
- Expecting exact float results  
- Using `==` for comparison  
- Assuming `int()` rounds  

---

## 🔹 Best Practices ✅
- Use `math.isclose()` for comparisons  
- Use `round()` for display only  
- Use `decimal` module for financial calculations  
- Be cautious in loops/aggregations  

---

## 🔹 Interview Takeaways 🎯
- Floats are **not exact**
- Based on **IEEE 754**
- Precision errors are unavoidable
- Always compare with tolerance

---

## 🔥 One-Line Summary
**Float = approximate decimal representation using binary with limited precision**
