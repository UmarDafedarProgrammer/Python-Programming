```python
# Binary approximation issue: 0.3 can't be stored exactly in binary.
# Python keeps the closest representable value instead.
# IEEE 754 double float: value = (-1)^sign * 1.mantissa * 2^(exponent - 1023)
# sign = 1 bit, exponent = 11 bits, mantissa = 52 bits.
0.1+0.2

print(0.1 + 0.2)  # Results in 0.30000000000000004
print(0.3)         # Prints 0.3, but the stored value is an approximation

# Computers only have a finite number of bits for floats
# (usually 52 bits in double precision), so values like 0.3
# get rounded to the nearest binary representation.
# That rounding is why you see 0.30000000000000004 instead of exactly 0.3.
```

## Understanding Float Storage vs Display

**Three Key Facts:**
1. Computers cannot store 0.3 exactly in binary
2. They store an approximation instead
3. print() hides the error, repr() shows it

```python
# EXAMPLE 1: What print() shows (cleaned up for readability)
print("What print() shows:")
print(0.3)  # You see: 0.3
```

```python
# EXAMPLE 2: What repr() shows (the TRUTH - what's actually stored)
print("What repr() shows (actual internal value):")
print(repr(0.3))  # You see: 0.29999999999999998
```

```python
# EXAMPLE 3: Using Decimal() to see the full truth
from decimal import Decimal
print("Decimal shows the EXACT internal bits:")
print(Decimal(0.3))  # Shows all the digits: 0.299999999999999988897769...
```

```python
# EXAMPLE 4: Why 0.1 + 0.3 = 0.4 (not 0.39...)
# Because they DON'T both round DOWN
from decimal import Decimal
print("0.1 is stored as:", Decimal(0.1))  # A bit HIGHER than 0.1
print("0.3 is stored as:", Decimal(0.3))  # A bit LOWER than 0.3
print("When added:")
print(repr(0.1 + 0.3))  # Result: 0.4
```

## Summary - The Three Layers

| Tool | What you see | Why |
|------|--------------|-----|
| `print(0.3)` | 0.3 | Python rounds for readability |
| `repr(0.3)` | 0.29999999999999998 | Shows what's really stored |
| `Decimal(0.3)` | 0.299999999999999988... | Shows all the binary bits converted back |

**Remember:** All three show the SAME value — just different levels of detail!

## Float built-in functions

Let's look at some built-in functions that are useful with floats, one at a time.

```python
# 1) float(): convert other values into a floating-point number
print(float(3))      # 3.0
print(float("2.75")) # 2.75
print(float(True))   # 1.0
```

```python
# 2) int(): convert a float to an integer by dropping the decimal part
print(int(3.9))      # 3
print(int(-3.9))     # -3
```

```python
# 3) round(): round to the nearest value, or to a specific number of decimals
print(round(2.75))    # 3
print(round(2.75, 1)) # 2.8
print(round(2.65, 1)) # 2.6 => Python rounds ties to the nearest even digit
print(round(2.5))     # 2
print(round(3.5))     # 4
print(round(-2.5))    # -2
print(round(2.345, 2))# 2.35 because it is not exactly stored as 2.345 instead 2.34500000000000000.....57578955

print(Decimal(2.345))

# print(round(3.456+6.456j)) complex does not define round method
```

```python
# abs() for floats
# abs() means absolute value.

# It always returns a non-negative number.
# It removes the sign, so it gives the distance from zero.

# How it works
# abs(3.2) → 3.2
# abs(-3.2) → 3.2
# abs(0.0) → 0.0

# Why it’s useful
# When you care about size, not direction.

# For example, the difference between 0.1 and 0.5:
# 0.1 - 0.5 is -0.4
# abs(0.1 - 0.5) is 0.4

# In words
# Think: “How far is this number from zero?”

# If it’s already positive, it stays the same.
# If it’s negative, it becomes positive.
```

```python
# float()
float(10)

# float() is the built-in Python function that converts a value into a floating-point number.

# What it does
# float(3) becomes 3.0
# float("2.75") becomes 2.75
# float(True) becomes 1.0
# float(False) becomes 0.0

# Why use it
# When you want a value to behave like a decimal number
# When you want division or math to use float rules
# When you read numeric text and need a number

# Important details
# It does not round; it only converts the type.
# If the string is not a valid number, Python raises ValueError.
# For some values, the float result is an approximation internally, because floats use binary storage.

# Good rule
# Use float() when you need a float value from:
# an integer
# a numeric string
# a boolean
# any object that can be converted to a number
```

```python
print(float(2.15))
print(float(6))
print(float("3.56"))
print(float(True))
print(float(False))
```

### `divmod()` for floats

`divmod(a, b)` returns two values:
- the quotient
- the remainder

For floats, it works like:
- quotient = `a // b`
- remainder = `a % b`

And Python guarantees:
- `a == b * quotient + remainder`
- `0 <= remainder < abs(b)` (if `b` is positive)

### Examples
- `divmod(7.5, 2.5)` → `(3.0, 0.0)`
  - because `7.5 = 2.5 * 3.0 + 0.0`
- `divmod(7.5, 2)` → `(3.0, 1.5)`
  - because `7.5 = 2 * 3.0 + 1.5`

### Why it is useful
- It gives both values at once, instead of writing two operations.
- It is handy when you need a quotient and remainder together.

### One more note
- If either argument is a float, the returned quotient is usually float.
- The remainder follows the same sign rules as Python’s `%`.

```python
divmod(14.6,5.6)
```

### `min()` and `max()`

These are built-in Python functions that find the smallest or largest value.

#### `min()`
- `min(a, b, c)` returns the smallest of the values.
- For floats, it compares numeric values normally.
- Example idea:
  - `min(2.5, 1.2, 3.7)` gives `1.2`

#### `max()`
- `max(a, b, c)` returns the largest of the values.
- For floats, it also works as expected.
- Example idea:
  - `max(2.5, 1.2, 3.7)` gives `3.7`

#### Useful patterns
- `min()` and `max()` are great when you have several numbers and want the smallest or biggest at once.
- They work with any comparable values, not just floats.

#### Why it matters for float learning
- With floats, `min()` and `max()` still look at the actual numeric values.
- So even if a float is stored as an approximation behind the scenes, these functions choose based on the value Python uses.


```python
min(2,3,4)
```

```python
max(3.4,.5,6,4.3,4.5)
```

### `ceil()` and `floor()`

These come from the `math` module, not built-in directly.

#### `math.ceil(x)`
- Returns the smallest integer that is **greater than or equal to** `x`
- In other words: round up
- Example:
  - `math.ceil(2.1)` → `3`
  - `math.ceil(2.0)` → `2`
  - `math.ceil(-2.1)` → `-2`

#### `math.floor(x)`
- Returns the largest integer that is **less than or equal to** `x`
- In other words: round down
- Example:
  - `math.floor(2.9)` → `2`
  - `math.floor(2.0)` → `2`
  - `math.floor(-2.1)` → `-3`

#### Important note
- For positive numbers:
  - `ceil` goes up
  - `floor` goes down
- For negative numbers:
  - `floor` goes more negative
  - `ceil` goes less negative

#### How to use


```python
import math
math.ceil(2.3)
math.floor(2.3)
```



These are useful when you need exact integer boundaries, for example:
- `ceil` to decide how many pages are needed
- `floor` to find the last whole item that fits
