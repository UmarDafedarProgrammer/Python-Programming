```python
#Q1. Find the factorial of 6

import math

def factorialN(n):
    if n < 0:
        raise ValueError("Please provide the positive Number")
    
    if type(n) != int: #isinstance(n,int) also can be used
        raise TypeError("Please provide integer values")
    
    if n==0 or n==1:
        return 1
    
    return factorialN(n-1) * n

    
```

```python
# recursive function
def gcd(a,b):

    if type(a) != int or type(b) != int:
        raise TypeError("should be integers")
    
    if a>=b:
        x,y = a,b
    else:
        x,y = b,a

    if y == 0:
        return x
        
    return gcd(y,x%y)
```

```python
gcd(25,5)
```

```python
# Q3. Find the LCM of 4,5 and 10

def lcm(a,b):
    if type(a) != int or type(b) != int:
        raise TypeError("should be integers")
    
    if a>=b:
        x,y = a,b
    else:
        x,y = b,a
    
    if y == 0:
        return 0
    
    return int(x*y/gcd(x,y))
    
```

```python
# Q4. Find the integer square root of 50

def sqrt(n):
    min = 0 
    max = n
    mid = 0


    while min <= max:
       
        mid = (min+max)//2

        if mid*mid == n:
            return mid
        elif mid*mid > n:
            max=mid -1
        elif mid*mid < n:
            min = mid + 1
    
    return max
    
```

```python
def factorial(n):

    if type(n) != int:
        raise TypeError("input should be integer")
    
    if n<0:
        raise ValueError("Inputn should be positive integer.")

    if n==0 or n==1:
        return 1

    return n * factorial(n-1)
```

```python
def combinations(a,b):

    return int(factorial(a)/(factorial(b)*factorial(a-b)))
```

```python
def permutations(a,b):
    return factorial(a)//factorial(a-b)
```

```python
print(math.comb(6,2))
print(combinations(6,2))
```

```python
print(math.perm(5,3))
print(permutations(5,3))
```

```python
# **Q8.** Check whether 144 is a perfect square (use logic, not sqrt directly)
# Q4. Find the integer square root of 50

def IsSqrt(n):
    min = 0 
    max = n
    mid = 0


    while min <= max:
       
        mid = (min+max)//2

        if mid*mid == n:
            return True
        elif mid*mid > n:
            max=mid -1
        elif mid*mid < n:
            min = mid + 1
    
    return False
    
```

```python
# **Q10.** Simplify the fraction 42 / 56 to lowest terms

def getlowestTerms(a,b):
    if a < b:
        x=b
        y=a
    else:
        x=a
        y=b
    
    d = gcd(x,y)

    return (x//d,y//d)
```

```python
getlowestTerms(56,42)
```
