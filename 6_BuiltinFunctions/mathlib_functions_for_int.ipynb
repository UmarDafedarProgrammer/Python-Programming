# 🔢 math.factorial() — Notes

## 📌 Definition

Factorial of a non-negative integer **n** is:

n! = n × (n−1) × (n−2) × ... × 1

---

## 🧠 Key Concept

* Represents **number of ways to arrange n distinct items**
* Used in **permutations, combinations, probability**

---

## 💻 Syntax

import math
math.factorial(n)

---

## ✅ Examples

math.factorial(0) = 1
math.factorial(1) = 1
math.factorial(5) = 120
math.factorial(7) = 5040

---

## ⚠️ Rules

1. Only accepts **non-negative integers**
2. Negative → ValueError
3. Float → TypeError

---

## ⭐ Important Identity

0! = 1

---

## ⚙️ Implementation

### 🔹 Iterative Approach

def factorial(n):
if n < 0:
raise ValueError("Negative not allowed")
result = 1
for i in range(1, n+1):
result *= i
return result

---

### 🔹 Recursive Approach

def factorial(n):
if n < 0:
raise ValueError("Negative not allowed")
if n == 0 or n == 1:
return 1
return n * factorial(n-1)

---

## 🚀 Properties

* Time Complexity: O(n)
* Space Complexity:

  * Iterative → O(1)
  * Recursive → O(n) (stack)
* Grows very fast (huge numbers)

---

## 🔥 Use Cases

1. Permutations → n!
2. Combinations → n! / (r! × (n−r)!)
3. Probability problems

---

## 🧠 Interview Tricks

### 1. Trailing Zeros in n!

Count number of 5s:

n//5 + n//25 + n//125 + ...

---

### 2. Growth Insight

10! = 3,628,800
20! ≈ 2.4 × 10^18

---

## ❗ Common Mistakes

* Using factorial for very large n
* Forgetting 0! = 1
* Passing float values

---

## ✅ Summary

* math.factorial(n) → product from n to 1
* Works only for non-negative integers
* Iterative preferred over recursion for large n
* Widely used in combinatorics


````python
### 🔹 Iterative Approach

def factorial(n):
    if n < 0:
        raise ValueError("Negative not allowed")
    result = 1
    for i in range(1, n+1):
        result *= i
    return result


### 🔹 Recursive Approach

def factorialRec(n):
    if n < 0:
        raise ValueError("Negative not allowed")
    if n == 0 or n == 1:
        return 1
    return n * factorialRec(n-1)
````

````python
def factorialN(n):
    if n<0:
        raise ValueError("Negative numbers are not allowed")
    if n == 1 or n == 0:
        return 1
    return n * factorialN(n-1)
````

# 🔢 math.gcd() — Notes

## 📌 Definition

GCD (Greatest Common Divisor) =
👉 Largest number that divides both numbers completely

---

## 🧠 Key Concept

* Also called **HCF (Highest Common Factor)**
* Used in:

  * Fractions simplification
  * Number theory
  * Cryptography
  * Algorithms

---

## 💻 Syntax

import math
math.gcd(a, b)

---

## ✅ Examples

math.gcd(12, 18) = 6
math.gcd(20, 30) = 10
math.gcd(7, 3) = 1
math.gcd(0, 5) = 5

---

## ⚠️ Rules

1. gcd(a, 0) = |a|
2. gcd(0, 0) = 0
3. Works with **multiple arguments**

math.gcd(12, 18, 24) = 6

---

## ⚙️ Implementation (Euclidean Algorithm)

### 🔹 Key Idea

gcd(a, b) = gcd(b, a % b)

---

### 🔹 Iterative Version

def gcd(a, b):
while b != 0:
a, b = b, a % b
return a

---

### 🔹 Recursive Version

def gcd(a, b):
if b == 0:
return a
return gcd(b, a % b)

---

## 🔍 Example (Step-by-step)

gcd(12, 18)

18 % 12 = 6
12 % 6 = 0

👉 Answer = 6

---

## 🚀 Properties

1. gcd(a, b) = gcd(b, a)
2. gcd(a, b) = gcd(b, a % b)
3. If gcd(a, b) = 1 → numbers are **coprime**

---

## 🔥 Use Cases

### 1. Simplifying Fractions

8/12 → divide by gcd(8,12)=4 → 2/3

---

### 2. LCM Formula

lcm(a, b) = (a × b) // gcd(a, b)

---

### 3. Coprime Check

math.gcd(a, b) == 1

---

## 🧠 Interview Tricks

### 🔹 Trick 1: Fast Computation

* Time Complexity: O(log n)
* Much faster than checking all divisors

---

### 🔹 Trick 2: Reduce multiple numbers

from functools import reduce
reduce(math.gcd, [12, 18, 24])

---

## ❗ Common Mistakes

* Trying brute force → slow
* Forgetting gcd(a,0)=a
* Not using modulo trick

---

## ✅ Summary

* math.gcd() finds largest common divisor
* Uses Euclidean Algorithm
* Very fast (logarithmic time)
* Extremely important for interviews


````python
import math
print(math.gcd(12,6))
print(math.gcd(6,12))
````

````python
math.gcd(3,7)
````

````python
math.gcd(3,0)
````

# 🔢 math.lcm() — Notes

## 📌 Definition

LCM (Least Common Multiple) =
👉 Smallest number that is divisible by both numbers

---

## 🧠 Key Concept

* Opposite of GCD
* Used in:

  * Fractions addition/subtraction
  * Scheduling problems
  * Cycles & periodic events

---

## 💻 Syntax

import math
math.lcm(a, b)

---

## ✅ Examples

math.lcm(4, 6) = 12
math.lcm(3, 5) = 15
math.lcm(7, 3) = 21
math.lcm(0, 5) = 0

---

## ⚠️ Rules

1. lcm(a, 0) = 0
2. Works with multiple arguments

math.lcm(4, 6, 8) = 24

---

## ⚙️ Implementation (Using GCD)

### 🔹 Formula

lcm(a, b) = (a × b) // gcd(a, b)

---

### 🔹 Iterative Version

def lcm(a, b):
import math
return (a * b) // math.gcd(a, b)

---

### 🔹 Multiple Numbers

from functools import reduce
import math

def lcm_list(nums):
return reduce(lambda x, y: (x*y)//math.gcd(x,y), nums)

---

## 🔍 Example (Step-by-step)

lcm(4, 6)

gcd(4, 6) = 2

lcm = (4 × 6) // 2 = 12

---

## 🚀 Properties

1. lcm(a, b) × gcd(a, b) = a × b
2. lcm(a, b) ≥ max(a, b)
3. If numbers are coprime → lcm = a × b

---

## 🔥 Use Cases

### 1. Fractions

LCM of denominators → common denominator

---

### 2. Scheduling Problems

“When will two events meet again?”

---

### 3. Cycles / Repeating Patterns

Used in time-based repetition problems

---

## 🧠 Interview Tricks

### 🔹 Trick 1: Use GCD

Never compute LCM directly → always use:
(a × b) // gcd(a, b)

---

### 🔹 Trick 2: Coprime Shortcut

If gcd(a, b) = 1 →
lcm = a × b

---

### 🔹 Trick 3: Avoid Overflow (important in some languages)

Use:
(a // gcd(a,b)) * b

---

## ❗ Common Mistakes

* Trying to find multiples manually
* Forgetting lcm(a,0)=0
* Not using gcd → inefficient

---

## ✅ Summary

* math.lcm() → smallest common multiple
* Always computed using gcd
* Very useful in fractions & cycles
* Important for problem solving


````python
math.lcm(4,5)
````

````python
math.gcd(4,5)
````

# 🔢 math.isqrt() — Notes

## 📌 Definition

Integer Square Root of n =
👉 Largest integer x such that:

x² ≤ n

---

## 🧠 Key Concept

* Returns **floor of square root**
* Avoids floating-point calculations
* Uses **binary search internally**

---

## 💻 Syntax

import math
math.isqrt(n)

---

## ✅ Examples

math.isqrt(10) = 3
math.isqrt(16) = 4
math.isqrt(25) = 5

---

## ⚠️ Rules

1. Only accepts **non-negative integers**
2. Negative → ValueError

---

## 🔍 Example: isqrt(25)

We need:
x² ≤ 25

Binary search steps:

left = 0, right = 25

mid = 12 → 144 ❌ → right = 11
mid = 5  → 25  ✅ → ans = 5, left = 6
mid = 8  → 64  ❌ → right = 7
mid = 6  → 36  ❌ → right = 5

Stop when left > right

👉 Final answer = 5

---

## ⚙️ Implementation (Binary Search)

def isqrt(n):
if n < 0:
raise ValueError("Negative not allowed")

```
left, right = 0, n  
ans = 0  

while left <= right:  
    mid = (left + right) // 2  

    if mid * mid <= n:  
        ans = mid  
        left = mid + 1  
    else:  
        right = mid - 1  

return ans  
```

---

## 🚀 Properties

* Time Complexity: O(log n)
* No precision errors (unlike sqrt)
* Works efficiently for large numbers

---

## 🔥 Use Cases

### 1. Perfect Square Check

if math.isqrt(n)**2 == n:
print("Perfect square")

---

### 2. Loop Optimization

for i in range(1, math.isqrt(n)+1):

---

### 3. Prime Checking

Check divisors only up to √n

---

## 🧠 Interview Insight

👉 This is **binary search on answer**

We search for:
largest x such that x² ≤ n

---

## ❗ Common Mistakes

* Using int(math.sqrt(n)) → precision issues
* Not understanding floor behavior
* Using linear loop → slow

---

## ✅ Summary

* math.isqrt(n) → integer square root
* Uses binary search
* Very efficient (log n)
* Important for optimization problems


````python
math.isqrt(10)
````

# 🔢 math.comb() — Notes

## 📌 Definition

Combination (nCr) =
👉 Number of ways to choose r items from n items (order does NOT matter)

---

## 🧠 Key Concept

* Selection, not arrangement
* Order is ignored

Example:
Choosing 2 from {A, B, C}

Valid:
AB, AC, BC → 3 ways

---

## 💻 Syntax

import math
math.comb(n, r)

---

## ✅ Examples

math.comb(5, 2) = 10
math.comb(6, 3) = 20
math.comb(10, 0) = 1
math.comb(10, 10) = 1

---

## ⚠️ Rules

1. n ≥ 0, r ≥ 0
2. r ≤ n
3. If r > n → ValueError

---

## ⭐ Formula

nCr = n! / (r! × (n−r)!)

---

## ⚙️ Implementation

### 🔹 Using Factorial

def comb(n, r):
import math
return math.factorial(n) // (math.factorial(r) * math.factorial(n-r))

---

### 🔹 Optimized (No large factorials)

def comb(n, r):
r = min(r, n-r)
result = 1

```
for i in range(r):  
    result = result * (n - i) // (i + 1)  

return result  
```

---

## 🔍 Example

math.comb(5, 2)

= 5! / (2! × 3!)
= (5×4×3×2×1) / ((2×1)(3×2×1))
= 10

---

## 🚀 Properties

1. nCr = nC(n−r)
2. nC0 = 1
3. nCn = 1
4. Symmetry → reduces computation

---

## 🔥 Use Cases

### 1. Selection Problems

Choose r items from n

---

### 2. Probability

Used in probability formulas

---

### 3. Subsets

Number of ways to pick elements

---

## 🧠 Interview Tricks

### 🔹 Trick 1: Reduce r

Always use:
r = min(r, n-r)

👉 Faster computation

---

### 🔹 Trick 2: Avoid factorial

Use iterative method → prevents overflow

---

### 🔹 Trick 3: Pascal Identity

nCr = (n-1)C(r-1) + (n-1)Cr

---

## ❗ Common Mistakes

* Confusing with permutation
* Using factorial for large n
* Forgetting symmetry

---

## ✅ Summary

* math.comb(n, r) → combinations
* Order does NOT matter
* Uses factorial logic
* Important for probability & counting problems


````python
n = [1,2,3,4,5,6]

math.comb(len(n),2)
````

# 🔢 math.perm() — Notes

## 📌 Definition

Permutation (nPr) =
👉 Number of ways to arrange r items from n items (order matters)

---

## 🧠 Key Concept

* Arrangement, not just selection
* Order is important

Example:
Choosing 2 from {A, B, C}

AB ≠ BA → both counted

---

## 💻 Syntax

import math
math.perm(n, r)

---

## ✅ Examples

math.perm(5, 2) = 20
math.perm(6, 3) = 120
math.perm(4, 4) = 24
math.perm(5, 0) = 1

---

## ⚠️ Rules

1. n ≥ 0, r ≥ 0
2. r ≤ n
3. If r > n → ValueError

---

## ⭐ Formula

nPr = n! / (n−r)!

---

## ⚙️ Implementation

### 🔹 Using Factorial

def perm(n, r):
import math
return math.factorial(n) // math.factorial(n-r)

---

### 🔹 Optimized (No full factorial)

def perm(n, r):
result = 1
for i in range(r):
result *= (n - i)
return result

---

## 🔍 Example

math.perm(5, 2)

= 5 × 4 = 20

---

## 🚀 Properties

1. nPr = nCr × r!
2. nP0 = 1
3. nPn = n!

---

## 🔥 Use Cases

### 1. Arrangements

Ranking, ordering, seating

---

### 2. Passwords / Codes

Order matters

---

### 3. Scheduling

Assigning positions/tasks

---

## 🧠 Interview Tricks

### 🔹 Trick 1: Think “order matters”

If positions matter → use permutation

---

### 🔹 Trick 2: Faster computation

Avoid factorial → use loop

---

### 🔹 Trick 3: Relation with combination

nPr = nCr × r!

---

## ❗ Common Mistakes

* Confusing with combination
* Using factorial unnecessarily
* Forgetting order matters

---

## ✅ Summary

* math.perm(n, r) → arrangements
* Order matters
* Faster than factorial method if optimized
* Very useful in real-world problems
