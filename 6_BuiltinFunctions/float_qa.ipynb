### How `math.isclose()` decides `True`

`math.isclose(a, b)` returns `True` when the difference between `a` and `b` is small enough under this rule:

`|a - b| <= max(rel_tol * max(|a|, |b|), abs_tol)`

So it uses:
- `rel_tol` for relative closeness
- `abs_tol` for absolute closeness

### Default values
- `rel_tol = 1e-9`
- `abs_tol = 0.0`

That means:
- Two values are allowed to differ by about 0.000000001 × the larger magnitude
- Or by at least `abs_tol` if you set it higher

### Example
For `math.isclose(0.1 + 0.2, 0.3)`:
- difference is around `5.55e-17`
- allowed tolerance is `1e-9 × 0.3 = 3e-10`
- since `5.55e-17 < 3e-10`, it returns `True`

### In plain terms
- If numbers are large, the allowed gap grows with size
- If numbers are tiny or near zero, the absolute tolerance matters
- If you want a specific threshold, set `rel_tol` or `abs_tol`
