#### Built-in Functions for Integers in Python

1. int() - Converts other data types into integers. Type Conversion Function

```python
print(int(5.9))
# print(int(3+4j)) int() argument must be a string, a bytes-like object or a real number, not 'complex'
print(int(True))
print(int(False))
print(int(-1.5))
print(int(-0.5))
print(int(0.5))
print(int("10"))
# print(int("10k")) invalid literal for int() with base 10: '10k'
print("-10")
print(" 10 ")
```

2. Mathematical Functions : abs() pow() round() divmod()

```python
#abs(): Returns a non-negative value
print(abs(10))
print(abs(-10))
print(abs(3.4))
print(abs(-3.4))

# Special case of complex number
print(abs(3+4j))
print(abs(3-4j)) #  (magnitude using √(a² + b²))

# print(abs([3,4,5]))  #bad operand type for abs(): 'list'
# list can not be passed as an argument to the abs()

# it will never take string as an input
# print(abs("10")) bad operand type for abs(): 'str'
```

```python
# round(inp, precision=0 places by default)
print(round(10)) 
print(round(10,1))
print(round(10.0,1))
print(round(3.4567))
print(round(-3.456))
print(round(4.567844566,3))
print(round(-4.567844566,3))
round(4.567)
round(4.5)
```

```python
# Banker's rounding - Round half to ecen
# Special case of rounding when the value is exactly 0.5/0.05/0.005 etc then it rounds to the nearest even number

# ✔ Python uses Banker’s rounding (half to even)
# ✔ Floating-point numbers are not exact
# ✔ So rounding can look “wrong” sometimes

# If you want accurate rounding (like finance) → use decimal:

print(round(1.35,1))
print(round(1.45,1))
print(round(1.355,2)) # this will print 1.35 due to binary approximation as it is actually stored as 1.35499999 in the memory
print(round(1.455,2))
```

```python
# one more special case

# When ndigits is negative:

# Value	Meaning
# -1	Round to nearest 10
# -2	Round to nearest 100
# -3	Round to nearest 1000

# In this also you will use bankers rounding when the value is exactly half

print(round(34,-1))
print(round(34,-2))
print(round(34,-3))
print(round(35,-1)) # Bankers rounding, round half to even
print(round(45,-1)) # Bankers rounding, round half to nearest even
print(round(350,-2))
print(round(450,-2)) # nearest even number, with rounding to nearest 100

# Negative digits → rounding happens on the integer side
# Python still applies normal rounding rules
# If exactly halfway → Banker’s rounding applies
```

```python
# pow() - pow(base, exponent, mod=None)

print(pow(2,3))
# 👉 2³ = 8

print(pow(2,-3))
# 💡 Important Points
# Negative exponent ⇒ result becomes a fraction
# Return type becomes float (even though inputs are integers)

print(pow(2.3,3))
print(pow(-2.3,3))

print(pow(-2,-3))
print(pow(-2,-2))
print(pow(-2,2))
print(pow(2,2))

pow(2, 3, 5)  # 3 (modulus version)
# pow(a,b,m)=(a**b)mod m

```

```python
# Division and remainder function
# (quotient, remainder) = divmod(dividend, divider)

# (a // b, a % b) returns

print(divmod(30,7))
print(divmod(-45,2))
```

6. format() - Converts a value into a formatted string based on a format specifier.

```python
# format(value, format_spec)

# What is format_spec?
# 👉 It is a mini-language (string pattern) that tells Python:
# “How should this value be displayed?”
# [[fill]align][sign][width][,/_][.precision][type]
# returns a string

print(format(10,'b')) # bninary 
print(format(10,'o')) # Octal numbers
print(format(10,'x')) # Hexdecimal with Little characters
print(format(10,'X')) # Hexadecimal with Capital Characters
print(format(10,'d')) # decimal 
```

8. type() and id()

```python
# type returns a class of data
i = 10
print(type(i))

s = 'string'
print(type(s))

f = 4.345
print(type(f))

b = True
print(type(b))

c = 'c'
print(type(c))

print(type(chr(65)))

print(type([1,2,3]))

print(type((1,2)))

print(type({1:2,2:3}))
```

```python
# id returns the unique memory address
print(id(i))
print(id(10))

print(id(f))
print(id(4.345))

print(id('c'))
print(id(c))

print(id('string'))
print(id(s))
```

```python
#has()

# “It returns an integer hash value used by Python to store and retrieve objects efficiently, especially in dictionaries and sets.”

print(hash(10))
print(hash('c'))
print(hash('string'))
print(hash(4.345))
print(hash(f))

# For integers, hash value = the integer itself


# Concept	Meaning
# hash()	Generates integer representation
# Purpose	Fast lookup
# For int	Same as value
# Works on	Immutable objects
# Used in	dict, set
```

#### ----------Arithmetic operations Along with Dunder methods---------------

```python
#addition +, __add__
print(3+4)

print((3).__add__(5))

a = 10 

print(a.__add__(30))

# 10.__add__(30) invalid decimal literal
# this considers 10. as floatinng point number, so make sure to include in parathtesis

# use class directly
print(int.__add__(10,30))
```

```python
## substraction, - , __sub__()

print(4-5)
print((4).__sub__(5))
print(int.__sub__(4,5))
```

```python
## multiplication, * , __mul__()

print(4*5)
print((4).__mul__(5))
print(int.__mul__(4,5))
```

```python
## floordivision, // , __floordiv__()
# integer value returned
print(53//5)
print((53).__floordiv__(5))
print(int.__floordiv__(53,5))
```

```python
## true division, // , __truediv__()
# Floating value returned

print(53/5)
print((53).__truediv__(5))
print(int.__truediv__(53,5))
```

```python
## modulus, // , __mod__()
# Returns a remainder

print(53%5)
print((53).__mod__(5))
print(int.__mod__(53,5))
```

```python
## exponential values, // , __pow__()

print(53**5)
print((53).__pow__(5))
print(int.__pow__(53,5))
```

```python
# <=, less than or equal to , __le__()

print(1<=2)
print((1).__le__(2))
print(int.__le__(1,2))
```

#### ----------Comparison operations Along with Dunder methods---------------

```python
## Bitwise operators, work on the bit level

# bitwise AND (&)
print(1 & 7) 

# 1 -> 0001
# &
# 7 -> 0111
# ----------
# 1 -> 0001 

print(7 & 7)
```

```python
# bitwise OR (|)

print(1 | 7)

# 1 -> 0001
# 7 -> 0111
#-----------
# 7 -> 0111
```

```python
# bitwise not (~)
~7
```

```python
## Left shift operator
# Operator	Meaning
# <<	Multiply
# >>	Divide

print(55<<2)
print(bin(55)[2:])
print(int(0b11011100))
print(55>>2)
print(int(0b001101))
```

```python
#Dunder methods
print((5).__and__(4))
print(int.__and__(4,5))

print((5).__or__(32))
print(int.__or__(32,5))

print((5).__xor__(45))
print(int.__xor__(45,5))

print((5).__lshift__(32))
print(int.__rshift__(32,5))

print(~20)
print(int.__invert__(20))
print((20).__invert__())
```
